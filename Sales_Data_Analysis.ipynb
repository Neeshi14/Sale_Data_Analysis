```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/Projects/ECOMM DATA.xlsx')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
total_sales = df['Sales'].sum()
total_sales
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date'].head()
```

```python
df.set_index('Order Date', inplace=True)
df.head()
```

```python
monthly_sales = df['Sales'].resample('M').sum()
monthly_sales
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.grid(True)
plt.show()
```

```python
best_sales_product = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
best_sales_product
```
